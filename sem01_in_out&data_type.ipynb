# Типы данных: целые числа — INT (INTEGER)


```python
print(11 + 2)  # сложение
print(11 - 2)  # вычитание
print(11 * 2)  # умножение
print(11 / 2)  # деление: результатом, как и на калькуляторе, будет дробное число.
```

Кроме того, компьютер умеет делить целые числа нацело, подсчитывая результат такого деления или остаток:

```python
print(11 // 2)  # деление нацело даст целое число в качестве результата
print(11 % 2)   # остаток от деления нацело
```

Ещё компьютер умеет возводить любые числа в степень:

```python
print(3 ** 2)   # 3 во второй степени = 9
print(3 ** 3)   # 3 в третьей степени = 27

print(3.5 ** 2) # 3.25 во второй степени = 12.25
```

# Типы данных: вещественные числа — FLOAT

**Вещественные числа** — число, в котором есть целая и дробная части. Целая и дробная части отделяются друг от друга точкой, а не запятой как в математике.

Даже если дробная часть числа равна нулю, все равно транслятор создаст в памяти вещественную переменную. Точка, как бы сигнал для транслятора, что необходимо создать именно вещественную переменную. 


С дробными числами тоже можно делать то же самое, что и с целыми:

```python
print(11.5 + 2)  # сложение
print(11.5 - 2)  # вычитание
print(11.5 * 2)  # умножение
print(11.5 / 2)  # деление

# делить нацело и брать остаток от деления — тоже можно:
print(11.5 // 2)    # деление нацело с дробным результатом
print(11.5 % 2)     # остаток от деления нацело 
```

Как объяснить компьютеру, с каким типом данных ему предстоит работать?

```python
a = input('Введите число: ')  # спросим два числа
b = input('и ещё одно: ')
print(a + b)
```

Почему так? Посмотрим, как понимает введённые числа компьютер с помощью функции `type()`:

```python
a = input('Введите число: ')  # спросим два числа
b = input('и ещё одно: ')
print(type(a), type(b))
print(a + b)
```

Изначально компьютер всё считает строками. Если нам нужен другой тип данных — нужно превратить строку в целое число, воспользовавшись функцией ``int()``, либо в дробное, с помощью функции ``float()``:


1. Спросим первое число (компьютер прочитает его строчкой).
2. Перезапишем переменную, превратив её из строки в целое число.
3. Спросим второе число, и сразу превратим его из строки в число, совместив две команды в одну.
4. Выведем результат.

```python
a = input('Введите число: ')   # спросим два числа
a = int(a)                     # перезапишем переменную a, указав, что было введено именно число
b = int(input('и ещё одно: ')) # для второго числа совместим две команды в одну
print(a + b)
```

**Задание 0.** Нам на корпоративе нужно посчитать вес кусков пиццы. Мы знаем, что пицца редко весит целое число килограмм и, обычно, на весах мы видим килограммы и граммы в виде дроби. Так же мы знаем, что пиццу получится порезать только на целое число кусков — на два, на три, ... на десять. p.s. И не забудем превратить вес каждого куска из числа в текст при выводе.

```python
# попросим компьютер узнать, на сколько кусков нам нужно поделить пиццу, с помощью функции int(), 
# сразу превратим количество кусков из текста в целое число
a = int(input('Введите количество кусков: ')) 

# попросим компьютер узнать, сколько весит торт и, с помощью функции float(),
# сразу превратим вес торта из текста в дробное число
b = float(input('Введите вес пиццы в килограммах: '))

# разделим вес пиццы на количество кусков
c = b / a 

# попробуем составить текст ответа, сложив его из фразы про вес одного куска, посчитанного веса и слова "килограмм"
d = 'Вес одного куска составит ' + str(c) + ' килограмм'  
print(d)
```

# Резюме за сегодня:
* Программы состоят из _команд_ — английских слов со скобками после них: print() — вывод, input() — ввод. Внутри скобок можно указать аргументы команды, т.е. то, что именно команда должна сделать. Например, что именно должна вывести команда print().
* Основные типы данных: _целые числа_ — тип, переменные которого могут содержать только целые числа (без дробной части, например: -2, -1, 0, 1, 2); _вещественные числа_ (или типы с плавающей точкой) — значения, имеющие дробную часть; _строка_ — тип представляющий из себя неизменяемую последовательность символов.
* Операции с целыми числами: сложение, вычитание, умножение, деление, деление нацело с дробным результатом и остаток от деления нацело.
* Операции с вещественными числами: такие же, как с целыми. Есть полезные функции для работы с подобным типом данных: _int()_ и _round()_.
