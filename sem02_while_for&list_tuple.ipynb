# Цикл while

![](https://i.ibb.co/w0dP6n0/Untitled-Frame-1-15.jpg)

`while` — один из самых универсальных циклов в Python, поэтому достаточно медленный.

Цикл `while` (в переводе — `пока`) позволяет выполнить одну и ту же последовательность действий, `пока` проверяемое условие истинно.


**Зачем**?

Как правило, цикл `while` используется, когда невозможно определить точное значение количества проходов исполнения цикла.

**Условие**?

Условие записывается до тела цикла и проверяется до выполнения тела цикла.

**Еще раз подробно**:

При выполнении цикла `while` сначала проверяется условие. Если оно ложно, то выполнение цикла прекращается и управление передается на следующую инструкцию после тела цикла `while`. Если условие истинно, то выполняется инструкция, после чего условие проверяется снова и снова выполняется инструкция. Так продолжается до тех пор, пока условие будет истинно. Как только условие станет ложно, работа цикла завершится и управление передастся следующей инструкции после цикла.

Синтаксис цикла `while` в простейшем случае выглядит так:

```
while <<условие>> :
    <<блок инструкций>>
    ```

### Задача № 1

Попробуем вывести все числа от 1 до заданного пользователем целого числа N с помощью цикла while.

````python
N = int(input()) # пользователь вводит число

i = 1            # вводим вспомогательный счетчик

while i <= N:    # проверяемое условие
    print(i)     # вывод числа на экран в теле цикла
    i = i + 1    # переходим к следующему числу, увеличиваем счетчик на 1 (Вариант 2: i+=1)
    
# заметим, что аналогично можно записывать и другие арифметические операции: например, -=
````

В этом примере переменная i внутри цикла изменяется от 1 до N. Такая переменная, значение которой меняется с каждым новым проходом цикла, называется счетчиком. 


! **Обратите внимание**, что код внутри цикла (тот, который мы хотим повторно выполнять), выделяется отступом.

### Задача № 2
Теперь попробуем посчитать из скольки цифр состоит натуральное число, введенное пользователем с клавиатуры с помощью цикла while.

````python
N = int(input())    # пользователь вводит число

length = 0          # вводим вспомогательный счетчик

while N > 0:        # проверяемое условие
    N //= 10        # целочисленное деление на 10
    length += 1     # переходим к следующему числу, увеличиваем счетчик на 1 
    
print(length)       # выводим результат
````

# Операторы break и continue

Циклами можно управлять с помощью операторов `break`, `continue`.

`break` внутри цикла позволяет прервать его выполнение и сразу же перейти к коду, который идет после цикла (либо завершить программу).

В этом случае мы можем написать сразу после цикла секцию `else` (синтаксис при этом такой же, как и в условиях).
Код, написанный после `else`, будет выполняться, если цикл завершился "естественным путем" (т.е. не был прерван с помощью `break`).

### Задача № 8

Рассмотрим пример задачи, которую можно решить с использованием break.

Пусть студент сдал 5 предметов во время сессии и мы хотим узнать, есть ли у него пересдачи.

ФОРМАТ ВВОДА

До пяти оценок от 1 до 10

ФОРМАТ ВЫВОДА

Если хотя бы одна из оценок меньше 4, завершаем программу и печатаем 'YES' (пересдачи есть).
Если все пять оценок больше 3, печатаем 'NO' (студент закрыл сессию без пересдач).


````python
i = 1
while i <= 5:
    mark = int(input('введите оценку: '))
    if mark < 4:
        print('YES')
        break
    i += 1
else: # else находится на том же уровне отступа, что и while, поэтому относится именно к циклу, а не к условию внутри цикла
    print ('NO')
````

Оператор `continue` позволяет сразу же перейти на новую итерацию цикла, не выполняя код, который написан внутри цикла ниже его.

Изменим условие задачи — теперь будем считать количество пересдач у студента.

````python
i = 1
retakes = 0
while i <= 5:
    mark = int(input("введите оценку: "))
    i += 1
    if mark >= 4: # если пересдачи нет, сразу же идем проверять переменную i, без увеличения переменной retakes
        continue
    retakes += 1
print("итого пересдач:", retakes)
````

Операторами `break` и `continue` не стоит злоупотреблять, это может ухудшить читаемость кода.

Например, в предыдущем примере мы бы справились и без `continue`:



````python
i = 1
retakes = 0
while i <= 5:
    mark = int(input("введите оценку: "))
    i += 1
    if mark < 4:
        retakes += 1
print("итого пересдач:", retakes)
````

# Цикл for

Мы уже умеем писать циклы с помощью оператора `while`, но теперь мы готовы познакомиться с еще одним способом.

Выведем список имен студентов по очереди. Для начала уже знакомым способом.

````python
students = ['Ivan Ivanov', 'Tatiana Sidorova', 'Maria Smirnova']

# len() возвращает длину строки

i = 0
while i < len(students):
    print(students[i])
    i += 1
````

В примере выше на каждом шаге цикл `for` достает очередной элемент из списка students и сохраняет его в переменную student.

**! Обратите внимание**, что переменную student нам не надо создавать заранее, она создается прямо во время работы цикла.

Получившийся код гораздо короче и проще варианта с использованием `while`.

````python
students = ['Ivan Ivanov', 'Tatiana Sidorova', 'Maria Smirnova']

for i in range(len(students)): # здесь не надо преобразовывать range в список или кортеж
    print(students[i])
````

### Функция `range()`

Пусть нам необходимо решить задачу: на вход подается число N, нужно сгенерировать список от 1 до N.

````python
# задачем изначальное положение итератора
i = 1

# создаем пустой элемент
myList = []

# вводим число
N = int(input())

# функция append() добавляет последующий элемент в конец списка
while i <= N:
    myList.append(i)
    i += 1

# выводим результат
print(myList)
````

Неплохо, но python предлагает нам готовое решение — встроенную функцию `range()`.

`range(N)` возвращает диапазон чисел от 0 до N - 1.

````python
myrange = range(10)
print(myrange[0], myrange[8], myrange[-1])
````

**Списки** `list`  — структура для хранения данных, последовательность которых необходимо запомнить. В большинстве же языков программирования используется термин **массив**. 


Список представляет собой последовательность элементов, пронумерованных от 0, как символы в строке. Список можно задать перечислением элементов списка в квадратных скобках, например, список можно задать так:

````python
# в списке numbers — 6 элементов, а именно: numbers[0] == 2, numbers[1] == 3, numbers[2] == 5, numbers[3] == 7, numbers[4] == 11, numbers[5] == 13. 
numbers = [2, 3, 5, 7, 11, 13]

#cписок elements состоит из 7 элементов, каждый из которых является строкой.
elements = ['первый', 'второй', 'третий', 'четвертый', 'пятый', 'шестой', 'седьмой']
````

> Также как и символы в строке, элементы списка можно индексировать отрицательными числами с конца. В отличие от строк, элементы списка можно изменять, присваивая им новые значения.

````python
numbers[-1] = 'тут поменяли'
numbers[-6] = 'и тут'
numbers
````

Длину также можно посмотреть с функцией _len()_.

````python
# создание списка

a = []  # заводим пустой список
# a = list() так тоже можно создать список
n = int(input())  # считываем количество элемент в списке
for i in range(n):  
    new_element = int(input())  # считываем очередной элемент
    a.append(new_element)  # добавляем его в список
    # последние две строки можно было заменить одной:
    # a.append(int(input()))
print(a)
````

````python
# тоже самое, но экономим переменную n

a = []
for i in range(int(input())):
    a.append(int(input()))
# вывод элементов списка, при этом будут выведены квадратные скобки вокруг элементов списка и запятые между элементами списка. 
print('такой себе вывод списка:', a)

# более притяный вывод списка
print('так-то лучше:')
for i in range(len(a)):
    print(a[i])

print('а так вообще отлично:')
for i in range(len(a)):
    print(a[i], end = ' ')
````

````python
# на вход подаётся строка
# 1 2 3
s = input()  # s == '1 2 3'
a = s.split()  # a == ['1', '2', '3']
# список будет состоять из строк, а не из чисел
print('список из строк:', a)

# теперь список из числе
for i in range(len(a)):
    a[i] = int(a[i])
print('список из чисел:', a)

# используя специальную магию питона — генераторы — то же самое можно сделать в одну строку:
# a = [int(s) for s in input().split()]
````

# Кортежи

**Кортежи** `tuple` — неизменяемый `list`.

````python
# создание кортежа

b = tuple()
b = ()

# создание кортежа из одного элементы

# так будет строка
c = ('s')
print(type(c))

# а так будет кортеж
b = ('s', )
print(type(b))
````

Операции с `tuple`:

Все операции над списками, **не изменяющие список** (сложение, умножение на число, методы index() и count() и т.п.). Можно также по-разному менять элементы местами и так далее.

````python
# p.s. гордость программистов на python — поменять местами значения двух переменных
a, b = b, a
````

Список —> хранит произвольные элементы.

Кортеж —> хранит произвольные элементы, неизменяемый. Защита от изменений, как намеренных (что плохо), так и случайных (что хорошо) и меньший размер. 

````python
a = [1, 2, 3, 4, 5, 6]
b = (1, 2, 3, 4, 5, 6)

print('размер списка =', a.__sizeof__())
print('размер кортежа =', b.__sizeof__())
````
